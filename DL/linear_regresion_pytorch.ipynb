### Pytorch教程：Linear Regression的Autograd实现

```python
import numpy as np
np.random.seed(0)
X = np.array([[1, 2], [2, 4], [3, 6], [4, 8], [6, 9]])
y = np.array([[8], [12], [16], [20], [25]])
# y = 2 * x1 + x2 + 4

w = np.random.rand(2, 1)
w
```

```python
w = torch.rand(2, 1, requires_grad=True, dtype=torch.float32)
w
```

```python
b = np.random.rand(1,1)
b
```

```python
iter_count = 10000
lr = 0.002
```

```python
def forword(x):
    return torch.matmul(x, w) + b
```

```python
def loss(y, y_pred):
    return ((y - y_pred) ** 2 / 2).sum()
```

```python
for i in range(iter_count):
    y_pred = forword(X)
    l = loss(y, y_pred)
    print(f'iter {i}, loss {l}')
    
    l.backward()
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad
        w.grad.zero_()
        b.grad.zero_()
        
```
